```python
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

# Suponiendo que 'restos_nlp' es tu DataFrame
categorias = restos_nlp[['name', 'stars_restaurant', 'address', 'cat_1', 'cat_2', 'cat_3', 'cat_4']].astype(str).values.tolist()

# Entrenar el modelo Word2Vec
model = Word2Vec(categorias, vector_size=100, window=5, min_count=1, workers=4)

# Función para obtener el vector promedio de un texto
def get_average_vector(words, model):
    valid_words = [word for word in words if word in model.wv]
    if not valid_words:
        return np.zeros(model.vector_size)
    return np.mean(model.wv[valid_words], axis=0)

# Añadir la columna 'vector' al DataFrame restos_nlp
restos_nlp['vector'] = restos_nlp.apply(lambda x: get_average_vector([x['name'], x['stars_restaurant'], x['address'], x['cat_1'], x['cat_2'], x['cat_3'], x['cat_4']], model), axis=1)

def find_similar_restaurants(name, num_similar_restaurants):
    restos_nlp.reset_index(drop=True, inplace=True)
    try:
        restaurant_index = restos_nlp.loc[restos_nlp['name'] == name].index[0]
    except IndexError:
        return [f'Disculpas, no pudimos encontrar el nombre "{name}". Por favor ingresa otro.']

    restaurant_vector = restos_nlp.loc[restaurant_index, 'vector']
    restaurant_vectors = np.stack(restos_nlp['vector'].values)

    cosine_sim = cosine_similarity([restaurant_vector], restaurant_vectors).flatten()
    restaurant_similarities = cosine_sim.argsort()[::-1][1:num_similar_restaurants+1]

    similar_restaurants = restos_nlp.loc[restaurant_similarities, ['name', 'address', 'stars_restaurant', 'cat_1', 'cat_2', 'cat_3', 'cat_4', 'mejor_reseña']]

    original_cat_1 = restos_nlp.loc[restos_nlp['name'] == name, 'cat_1'].values[0]
    print(f'Restaurante original: {name}, categoría principal: {original_cat_1} \n')
    print('---')
    print(f'Restaurantes similares:\n')

    for i, restaurant in similar_restaurants.iterrows():
        print(f'Nombre: {restaurant[0]}')
        print(f'Puntuación: {restaurant.stars_restaurant}')
        print(f'Dirección: {restaurant.address}')
        print(f'Categoría 1: {restaurant.cat_1}')
        print(f'Categoría 2: {restaurant.cat_2}')
        print(f'Categoría 3: {restaurant.cat_3}')
        print(f'Categoría 4: {restaurant.cat_4}')
        print(f"\nMejor Reseña: {restaurant.mejor_reseña}")
        print('---')
```

```python
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

# Suponiendo que 'restos_nlp' es tu DataFrame
categorias = restos_nlp[['name', 'stars_restaurant', 'cat_1', 'cat_2', 'cat_3', 'cat_4']].astype(str).values.tolist()
model = Word2Vec(categorias, vector_size=100, window=5, min_count=1, workers=4)

# Función para obtener el vector promedio de un texto
def average_vector(words, model):
    valid_words = [word for word in words if word in model.wv]
    if not valid_words:
        return np.zeros(model.vector_size)
    return np.mean(model.wv[valid_words], axis=0)

# Añadir la columna 'vector' al DataFrame restos_nlp
restos_nlp['vector'] = restos_nlp.apply(lambda x: average_vector(
    [x['name'], x['stars_restaurant'], x['cat_1'], x['cat_2'], x['cat_3'], x['cat_4']], model), axis=1)

# Función para obtener las calificaciones de un usuario
def get_user_ratings(user_id):
    return stars[stars['user_id'] == user_id].sort_values(by='stars_review', ascending=False)

# Función para recomendar restaurantes similares basados en contenido
def recommend_similar_restaurants(user_id, num_similar_restaurants=3):
    user_ratings = get_user_ratings(user_id)

    if user_ratings.empty:
        return f"El usuario '{user_id}' no ha puntuado ningún restaurante."

    # Obtener los restaurantes mejor puntuados por el usuario
    top_rated_restaurants = user_ratings[user_ratings['stars_review'] == user_ratings['stars_review'].max()]

    if top_rated_restaurants.empty:
        return f"El usuario '{user_id}' no tiene calificaciones válidas."

    recommendations = []
    top_rated_details = []

    for _, top_rated in top_rated_restaurants.iterrows():
        top_rated_restaurant_id = top_rated['business_id']
        top_rated_info = restos_nlp.loc[restos_nlp['business_id'] == top_rated_restaurant_id, ['name', 'stars_restaurant', 'cat_1', 'cat_2', 'cat_3', 'cat_4']]
        top_rated_details.append(top_rated_info.iloc[0])

        # Obtener el índice del restaurante mejor puntuado
        try:
            restaurant_index = restos_nlp.loc[restos_nlp['business_id'] == top_rated_restaurant_id].index[0]
        except IndexError:
            continue

        # Calcular similitud del coseno
        restaurant_vector = restos_nlp.loc[restaurant_index, 'vector']
        restaurant_vectors = np.stack(restos_nlp['vector'].values)
        cosine_sim = cosine_similarity([restaurant_vector], restaurant_vectors).flatten()

        # Obtener los restaurantes más similares, excluyendo el restaurante ingresado
        sorted_indices = np.argsort(cosine_sim)[::-1]
        restaurant_similarities = sorted_indices[1:num_similar_restaurants + 1]

        similar_restaurants = restos_nlp.iloc[restaurant_similarities][['name', 'stars_restaurant', 'address', 'cat_1', 'cat_2', 'cat_3', 'cat_4']]

        for _, restaurant in similar_restaurants.iterrows():
            # Aquí corregimos para acceder correctamente a 'mejor_reseña'
            restaurant_details = restos_nlp.loc[restos_nlp['name'] == restaurant['name'], ['mejor_reseña']].iloc[0]
            recommendations.append({
                'name': restaurant['name'],
                'stars_restaurant': restaurant['stars_restaurant'],
                'address': restaurant['address'],
                'cat_1': restaurant['cat_1'],
                'cat_2': restaurant['cat_2'],
                'cat_3': restaurant['cat_3'],
                'cat_4': restaurant['cat_4'],
                'mejor_reseña': restaurant_details['mejor_reseña']
            })

    # Imprimir las características de los restaurantes mejor puntuados por el usuario
    print(f"Restaurantes mejor puntuados por el usuario '{user_id}':\n")
    for details in top_rated_details:
        print(f"Nombre: {details['name']}")
        print(f"Estrellas: {details['stars_restaurant']}")
        print(f"Categoría 1: {details['cat_1']}")
        print(f"Categoría 2: {details['cat_2']}")
        print(f"Categoría 3: {details['cat_3']}")
        print(f"Categoría 4: {details['cat_4']}")
        print('---')

    # Imprimir las recomendaciones
    print(f"Restaurantes similares recomendados para el usuario '{user_id}':\n")
    for restaurant in recommendations:
        print(f"Nombre: {restaurant['name']}")
        print(f"Estrellas: {restaurant['stars_restaurant']}")
        print(f"Dirección: {restaurant['address']}")
        print(f"Categoría 1: {restaurant['cat_1']}")
        print(f"Categoría 2: {restaurant['cat_2']}")
        print(f"Categoría 3: {restaurant['cat_3']}")
        print(f"Categoría 4: {restaurant['cat_4']}")
        print(f"\nMejor Reseña: {restaurant['mejor_reseña']}")
        print('---')

# Solicitar un user_id al usuario y recomendar restaurantes
user_id = input("Ingresa tu user ID: ")
recommend_similar_restaurants(user_id)
```

###Modelo de recomendación por contenido: palabra clave y estrellas mínimas

```python
def recommend_restaurants(keyword, restos_nlp, min_stars=0, num_restos_to_recommend=None):
    # Filtrar los restaurantes que contienen la palabra clave en alguna de las categorías
    keyword_lower = keyword.lower()
    filtered_restos = restos_nlp[
        restos_nlp.apply(lambda x: keyword_lower in x['cat_1'].lower() or
                                   keyword_lower in x['cat_2'].lower() or
                                   keyword_lower in x['cat_3'].lower() or
                                   keyword_lower in x['cat_4'].lower(), axis=1)
    ]

    # Filtrar por estrellas promedio mínimas
    filtered_restos = filtered_restos[filtered_restos['stars_restaurant'] >= min_stars]

    # Ordenar por estrellas promedio
    filtered_restos = filtered_restos.sort_values(by='stars_restaurant', ascending=False)

    # Limitar el número de restaurantes recomendados
    if num_restos_to_recommend is not None:
        filtered_restos = filtered_restos.head(num_restos_to_recommend)

    if filtered_restos.empty:
        return "No se encontraron restaurantes que coincidan con los criterios."

    # Include 'mejor_reseña' in the returned columns
    return filtered_restos[['name', 'address', 'stars_restaurant', 'cat_1', 'cat_2', 'cat_3', 'cat_4', 'mejor_reseña']]

# Ejemplo de uso
keyword = input("Ingresa la palabra clave para buscar en las categorías: ")
min_stars = float(input("Ingresa la calificación mínima de estrellas (0-5): "))
num_restos_to_recommend = int(input("Ingresa la cantidad de restaurantes a recomendar: "))

recommended_restos = recommend_restaurants(keyword, restos_nlp, min_stars, num_restos_to_recommend)

if isinstance(recommended_restos, str):
    print(recommended_restos)
else:
    print("Restaurantes recomendados:")
    print("---")
    for _, row in recommended_restos.iterrows():
        print(f"Nombre: {row['name']}")
        print(f"Puntuación: {row['stars_restaurant']}")
        print(f"Dirección: {row['address']}")
        print(f"Categoría 1: {row['cat_1']}")
        print(f"Categoría 2: {row['cat_2']}")
        print(f"Categoría 3: {row['cat_3']}")
        print(f"Categoría 4: {row['cat_4']}")
        print(f"\nMejor Reseña: {row['mejor_reseña']}")
        print('---')
```

### ¿Para qué sirve este código?

El sistema intenta proporcionar recomendaciones personalizadas basadas en la información proporcionada por el usuario, ya sea a través de su historial de reseñas, el nombre de un restaurante conocido, o mediante la búsqueda por palabra clave y calificación mínima.

### ¿Cómo funciona?
Pide al usuario que ingrese su user_id.  
Si el user_id existe en el DataFrame stars, es decir, realizo reseñas previamente, le ofrece al usuario seleccionar el metodo de busqueda para la recomendacion.
  - Según mi historial.
  - Restaurantes similares.
  - Por categoria y estrellas minimas.

En caso de que el id no haya realizado reseñas previamente, solo tiene dos opciones de busqueda:
  - Por Restaurantes similares.
  - Por categoria y estrellas minimas.

En cualquiera de los casos, si el nombre del restaurante no existe, pide una palabra clave y una calificación mínima de estrellas, y llama a recommend_restaurants para recomendar restaurantes basados en la categoria ingresada y el numero minimo de estrellas. Además para que el usuario tenga más información al respecto sobre el local le proporcionamos la reseña más representativa del mismo.


###Modelo de recomendación colaborativo: usuarios


```python
user_item_matrix = stars.pivot_table(index='user_id', columns='business_id', values='stars_review', fill_value=0)
```

```python
# Calculate cosine similarity (or Pearson correlation if preferred)
user_similarity = cosine_similarity(user_item_matrix)

# Function to find similar users
def find_similar_users(user_id, num_similar_users=5):
  user_index = user_item_matrix.index.get_loc(user_id)
  similar_users_indices = user_similarity[user_index].argsort()[::-1][1:num_similar_users+1]
  return user_item_matrix.index[similar_users_indices]
```

```python
threshold = 3
positive_ratings = {}
for user_id in user_item_matrix.index:
  positive_ratings[user_id] = stars[(stars['user_id'] == user_id) & (stars['stars_review'] >= threshold)]['business_id'].tolist()

untried_restaurants = {}
for user_id in user_item_matrix.index:
  untried_restaurants[user_id] = []
  similar_users = find_similar_users(user_id)
  for similar_user in similar_users:
    for restaurant in positive_ratings[similar_user]:
      if restaurant not in positive_ratings[user_id]:
        untried_restaurants[user_id].append(restaurant)

users_with_untried = [user_id for user_id, restaurants in untried_restaurants.items() if restaurants]
```

```python
import random

def recommend_restaurants_with_reasons(user_id, threshold=3, num_similar_users=10, num_recommendations=5):
  similar_users = find_similar_users(user_id, num_similar_users)
  recommendations = []

  # Calculate average rating for target user
  user_avg_rating = stars[stars['user_id'] == user_id]['stars_review'].mean()

  for similar_user in similar_users:
    # Calculate average rating for similar user
    similar_user_avg_rating = stars[stars['user_id'] == similar_user]['stars_review'].mean()

    rated_restaurants = stars[
        (stars['user_id'] == similar_user) &
        (stars['stars_review'] >= threshold) &
        (~stars['business_id'].isin(stars[stars['user_id'] == user_id]['business_id']))
    ]

    for _, restaurant in rated_restaurants.iterrows():
      restaurant_name = restos[restos['business_id'] == restaurant['business_id']]['name'].values[0]
      recommendations.append({
          'restaurant_name': restaurant_name,
          'restaurant_stars': restaurant['stars_review'],
          'similar_user_id': similar_user,
          'similar_user_avg_rating': similar_user_avg_rating,
          'your_avg_rating': user_avg_rating
      })

    if len(recommendations) >= num_recommendations:
      break

  return recommendations[:num_recommendations]

# Select a random user with untried restaurants
random_user_id = random.choice(users_with_untried)

# Generate recommendations with reasons
recommendations_with_reasons = recommend_restaurants_with_reasons(random_user_id)

# Display recommendations with reasons
if recommendations_with_reasons:
  print(f"Recomendaciones para el usuario {random_user_id}:")
  for recommendation in recommendations_with_reasons:
    print(f"- {recommendation['restaurant_name']} ({recommendation['restaurant_stars']} estrellas)")
    print(f"  Recomendado porque al usuario {recommendation['similar_user_id']} (con rating promedio de: {recommendation['similar_user_avg_rating']:.2f}) le gustó, "
          f"y tu rating promedio es {recommendation['your_avg_rating']:.2f}.")
else:
  print(f"No hay recomendaciones de restaurantes que el usuario {random_user_id} no haya probado.")
```

Este es un sistema de recomendación colaborativo basado en usuarios, es decir, recomienda restaurantes similares en cuanto al contenido que ven usuarios que se parecen a vos por cómo puntúan. Es decir, si yo puntúo en promedio 4, me va a traer los restos que puntuaron otros usuarios que puntuan como yo.

El threshold de 3 es para que las recomendaciones sean de neutras a positivas (ignorando las negativas)

También, el sistema me recomendará restaurantes que el user_id ingresado no haya ido/puntuado.

Por esta razón es que puede pasar que ingrese un user_id y no haya recomendaciones.

```python
from pyspark.sql import SparkSession
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.feature import StringIndexer

# Crear una sesión de Spark
spark = SparkSession.builder \
    .appName("ALS Restaurant Recommendation") \
    .config("spark.executor.memory", "2g") \
    .config("spark.driver.memory", "2g") \
    .config("spark.sql.shuffle.partitions", "10") \
    .getOrCreate()

# Supongamos que 'stars' es tu DataFrame de pandas
# Convertir DataFrame de pandas a DataFrame de Spark
df_spark = spark.createDataFrame(stars)

# Convertir 'user_id' y 'business_id' a índices numéricos
indexer_user = StringIndexer(inputCol="user_id", outputCol="user_id_index")
indexer_business = StringIndexer(inputCol="business_id", outputCol="business_id_index")

df_indexed = indexer_user.fit(df_spark).transform(df_spark)
df_indexed = indexer_business.fit(df_indexed).transform(df_indexed)

# Convertir las columnas de índice a tipo entero
df_indexed = df_indexed.withColumn("user_id_index", df_indexed["user_id_index"].cast("integer"))
df_indexed = df_indexed.withColumn("business_id_index", df_indexed["business_id_index"].cast("integer"))

# Dividir los datos en entrenamiento y prueba
(train_data, test_data) = df_indexed.randomSplit([0.8, 0.2])

# Crear el modelo ALS
als = ALS(userCol="user_id_index", itemCol="business_id_index", ratingCol="stars_review", coldStartStrategy="drop")

# Construir una cuadrícula de parámetros (simplificada)
param_grid = (ParamGridBuilder()
              .addGrid(als.rank, [10, 50])
              .addGrid(als.maxIter, [10])
              .addGrid(als.regParam, [0.1])
              .build())

# Definir el evaluador
evaluator = RegressionEvaluator(metricName="rmse", labelCol="stars_review", predictionCol="prediction")

# Configurar la validación cruzada
crossval = CrossValidator(estimator=als,
                          estimatorParamMaps=param_grid,
                          evaluator=evaluator,
                          numFolds=3)

# Entrenar el modelo con validación cruzada
try:
    cv_model = crossval.fit(train_data)
except Exception as e:
    print(f"Error durante el entrenamiento del modelo: {e}")
    spark.stop()
    raise

# Obtener el mejor modelo
best_model = cv_model.bestModel

# Evaluar el mejor modelo
predictions = best_model.transform(test_data)
rmse = evaluator.evaluate(predictions)
print(f"Root Mean Squared Error (RMSE) del mejor modelo = {rmse}")

# Detener la sesión de Spark
spark.stop()
```
